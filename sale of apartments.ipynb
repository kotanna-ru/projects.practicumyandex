Пропуски в столбцах  **"living_area"** и **kitchen_area**  и **"balcony"**

С юридической точки зрения апартаменты признаются нежилой недвижимостью. Мы можем заполнить пропуски  в столбцах с жилой площадью и площадью кухни нулевыми значения. Так же поступим и с квартирами со свободной планировкой(open_plan) в отношении площади кухни

```python
for i in data['is_apartment']:
    data.loc[(data['is_apartment'] == True) & (data['living_area'].isna()), 'living_area'] = 0
    data.loc[(data['is_apartment'] == True) & (data['kitchen_area'].isna()), 'kitchen_area'] = 0
for i in data['open_plan']:
    data.loc[(data['open_plan'] == True) & (data['kitchen_area'].isna()), 'kitchen_area'] = 0
```

Все эти показатели коррелируют с показателями общей площади. А она в свою очередь с показателем - количество комнат.

Оставшиеся пропуски мы заменим медианой  жилой площади/ кухни и балконов в соотношении с количеством комнат в  квартире 

Выведем на экран уникальные значения количества комнат в квартире.

```python
sorted(data['rooms'].unique())
```

Создадим функцию part_apart, которая заполнит пропуски  в столбце медианой в соответствии с количеством комнат

```python
part=['living_area','kitchen_area','balcony']
def part_apart(part):
    for member in part:
        for room in data['rooms'].unique():
            data.loc[(data['rooms'] == room) & (data[member].isna()), member] =\
            data.loc[(data['rooms'] == room), member].median()
```

Заполним пропуски в нашей таблице применив функцию part_apart. 

```python
part_apart(part)
```

Продолжим искать дубликаты. Создадим функцию, которая из оставшихся уникальных значений возьмет только последнее слово в названии, и посчитает сколько раз оно повторилось

```python
locality_name=sorted(data['locality_name'].unique())
name_unique=[]
for i in locality_name:
    name=i.split()
    name_unique.append(name[-1])
sorted(name_unique)  
name_unique=pd.DataFrame(name_unique)
name_unique[0].value_counts()
```

Получилось 295 уникальных значений, уберем те которые не повторяются

```python
name_unique_min=pd.DataFrame(name_unique[0].value_counts()).reset_index()
name_unique_min=name_unique_min[name_unique_min[0]>1]
name_unique_min['index']
```

Таких названий 25. Дальше добавим в срез таблицы, где есть все населённые пункты столбик куда запишем найденные нами 25 значений, только в те строки у которых названия населённых пунктов оканчиваются на одно из этих значений

```python
def categorize(row):
    for i in name_unique_min['index']:
        if i in row:
            return i
            
data_name_loc=data[['locality_name']]
data_name_loc['category'] =data_name_loc['locality_name'].apply(categorize)

```

Сгруппируем. Выведем на экран. И отберем те названия которые не относятся к одному населенному пункту

```python
data_name_loc_gr=data_name_loc.groupby(['category','locality_name']).count()

data_name_loc_gr
```

```python
deled=['Бегуницы', 'Бор', 'Горка','Колпаны','Ладога','Приладожский','Рахья,''Романовка','Ропша']
for i in name_unique_min['index']:
    if i not in deled:
        for j in data['locality_name']:
            if j.endswith(i):
                data['locality_name'] = data['locality_name'].str.replace(j, 'поселок '+i)
```

#### Изучим как стоимость объектов зависит от расстояния до центра города.

```python
#сделаем срез столбцов с данными о стоимости, населенным пунктом и расстоянии до центра в км.
data_cut_centers=data[['last_price','locality_name','city_centers_km']]
data_cut_centers
```

```python
data_cut_sp=data_cut_centers.query('locality_name =="Санкт-Петербург"')
data_cut_sp['price_km']=data_cut_sp['last_price']/data_cut_sp['city_centers_km']
data_cut_sp=data_cut_sp.dropna() #удалим пустые значения
data_cut_sp['price_km']=round(data_cut_sp['price_km'],2)
data_cut_sp
```

Разобьем расстояние до центра на 4 категории центр, близко, далеко, очень далеко.

```python
categorize_4= ['центр', 'близко', 'далеко', 'очень далеко']

data_cut_sp['categorize'] = pd.qcut(data_cut_sp['city_centers_km'],q=4, labels=categorize_4)
data_cut_sp.head()
```

```python
data_cut_sp['categorize'].value_counts()
```

Выведем на экран среднею стоимость 1 км в каждой категории и среднее расстояние

```python
data_cut_sp.pivot_table(index='categorize', values=['price_km','city_centers_km'], aggfunc='median')
```

Построим график

```python
data_cut_sp.pivot_table(index='categorize', values='price_km', aggfunc='median').plot.bar();
```
